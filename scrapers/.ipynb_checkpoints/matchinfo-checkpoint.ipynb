```python
import cloudscraper
import pandas as pd
```

```python
url = "https://fbref.com/en/comps/9/schedule/Premier-League-Scores-and-Fixtures"

scraper = cloudscraper.create_scraper()

html = scraper.get(url).text

tables = pd.read_html(html, attrs={"id":"sched_2025-2026_9_1"})
df = tables[0]
```

```python
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
today = pd.Timestamp.today().normalize()
df = df[df["Date"] <= today].copy()
df = df.drop(columns=["Wk","Day", "Time", "Attendance", "Venue", "Referee", "Match Report", "Notes"])
df.to_csv("../CSV_files/fixtures.csv", index=False)
```
