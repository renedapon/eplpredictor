```python
import cloudscraper
import pandas as pd

url = "https://fbref.com/en/comps/9/Premier-League-Stats"

scraper = cloudscraper.create_scraper()

html = scraper.get(url).text

tables = pd.read_html(html, attrs={"id": "stats_squads_standard_for"})
df = tables[0]
```

```python
df.columns = ['_'.join(col).strip() for col in df.columns.values]
df.head()
```

```python
df = df.rename(columns={
    "Unnamed: 0_level_0_Squad": "Squad",
    "Unnamed: 1_level_0_# Pl": "Players",
    "Unnamed: 2_level_0_Age": "Avg_player_age",
    "Unnamed: 3_level_0_Poss": "Possession"
})
df.head()
```

```python
df.to_csv("squad_standard_stats_2025-2026.csv", index=False)
```
