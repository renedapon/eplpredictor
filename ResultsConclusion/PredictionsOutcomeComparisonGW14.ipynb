**Comparsion Table**

```python
import matplotlib.pyplot as plt

plt.style.use("dark_background")
plt.rcParams['figure.figsize'] = (10, 5)
plt.rcParams['font.size'] = 13

fixtures = [
    "Fulham vs Man City",
    "Bournemouth vs Everton",
    "Newcastle Utd vs Tottenham",
    "Brighton vs Aston Villa",
    "Burnley vs Crystal Palace",
    "Wolves vs Nott'ham Forest",
    "Arsenal vs Brentford",
    "Leeds Utd vs Chelsea",
    "Liverpool vs Sunderland",
    "Manchester Utd vs West Ham",
    "Accuracy"
]

model_pred = [
    "Draw", "Home win", "Home win", "Home win", "Away win",
    "Away win", "Home win", "Away win", "Home win", "Home win",
    "3"
]

bet365_pred = [
    "Away win", "Home win", "Home win", "Home win", "Away win",
    "Away win", "Home win", "Away win", "Home win", "Home win",
    "4"
]

actual = [
    "Away win", "Away win", "Draw", "Away win", "Away win", "Away win",
    "Home win", "Home win", "Draw", "Draw",
    "-"
]

table_data = list(zip(fixtures, model_pred, bet365_pred, actual))

fig, ax = plt.subplots()
ax.axis("off")

table = ax.table(
    cellText=table_data,
    colLabels=["Fixture", "Our Model Prediction", "Bet365 Prediction", "Actual Result"],
    cellLoc="center",
    loc="center"
)


for (row, col), cell in table.get_celld().items():

    if row == 0:
        cell.set_facecolor("#333333")
        cell.set_text_props(color="white", weight="bold")
        continue

    cell.set_facecolor("#1E1E1E")
    cell.set_text_props(color="white")

    fixture_idx = row - 1 

    true = actual[fixture_idx]

    if col == 1:
        pred = model_pred[fixture_idx]
        if true == "-":
            cell.set_facecolor("#1E1E1E")
        elif pred == true:
            cell.set_facecolor("#0B6623") 
            cell.set_text_props(color="white", weight="bold")
        else:
            cell.set_facecolor("#8B0000")
            cell.set_text_props(color="white", weight="bold")

    if col == 2:
        pred = bet365_pred[fixture_idx]
        if true == "-":
            cell.set_facecolor("#1E1E1E")
        elif pred == true:
            cell.set_facecolor("#0B6623") 
            cell.set_text_props(color="white", weight="bold")
        else:
            cell.set_facecolor("#8B0000")
            cell.set_text_props(color="white", weight="bold")

table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1, 2)

plt.tight_layout()
plt.savefig("gw14_predictions_table.png", dpi=300, bbox_inches="tight", facecolor=fig.get_facecolor())
plt.show()
```

**Interpretation of the Prediction Comparison Table**

This table compares the most likely match outcome predicted by our model with Bet365’s favourite outcome for each Gameweek 14 fixture. In almost all the cases, both systems choose the same result, indicating that our model generally aligns with bookmaker expectations.

Despite this agreement, Premier League matches remain highly unpredictable, and even the most likely outcome often fails to materialise. This explains why both our model and Bet365 miss several fixtures. The underlying probabilities may be similar, but the real-world match outcomes frequently defy expectations due to the dynamic and volatile nature of football.

It is also important to note that evaluating accuracy on a single gameweek is not reliable.
A meaningful assessment requires testing the model across many gameweeks or entire seasons, where randomness averages out and long-term patterns become visible. Only then can we determine the true predictive power and stability of the model.
